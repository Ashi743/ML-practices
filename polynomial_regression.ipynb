```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

def Polynomial_regression(x, y, degree, x_test):
    # Create a new set of x values for plotting the polynomial curve
    x_new = np.linspace(-3, 3, 200).reshape(-1, 1)

    # Define the pipeline for polynomial regression
    pipeline = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('lr', LinearRegression())
    ])

    # Fit the pipeline on the training data
    pipeline.fit(x, y)

    # Predict using the pipeline on the new x values and x_test
    y_new = pipeline.predict(x_new)
    y_pred1 = pipeline.predict(x_test)

    # Plot the original data points
    plt.scatter(x, y, label="Training points")

    # Plot the polynomial regression curve
    plt.plot(x_new, y_new, c='r', label=f"Polynomial degree {degree}")

    # Plot the predictions on the test data
    plt.scatter(x_test, y_pred1, c='b', label="Test predictions", marker='x')

    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Polynomial Regression (degree {degree})")
    plt.axis([-5, 5, 0, 10])
    plt.show()


```

```python
# 3D polynomial regression
x = 7 * np.random.rand(100, 1) - 2.8
y = 7 * np.random.rand(100, 1) - 2.8

z = x**2 + y**2 + 0.2*x + 0.2*y + 0.1*x*y +2 + np.random.randn(100, 1)
# z = x^2 + y^2 + 0.2x + 0.2y + 0.1xy + 2
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter_3d(df, x=x.ravel(), y=y.ravel(), z=z.ravel())
fig.show()
```

```python
lr = LinearRegression()
lr.fit(np.array([x,y]).reshape(100,2),z)

x_input = np.linspace(x.min(), x.max(), 10)
y_input = np.linspace(y.min(), y.max(), 10)
xGrid, yGrid = np.meshgrid(x_input,y_input)

final = np.vstack((xGrid.ravel().reshape(1,100),yGrid.ravel().reshape(1,100))).T

z_final = lr.predict(final).reshape(10,10)
```

```python
import plotly.graph_objects as go

fig = px.scatter_3d(df, x=x.ravel(), y=y.ravel(), z=z.ravel())

fig.add_trace(go.Surface(x = x_input, y = y_input, z =z_final ))

fig.show()
```
