```python
orders["date_datetime"]= pd.to_datetime(orders["date"])
```

```python
orders
```

```python
orders["year"]=  orders["date_datetime"].dt.year
orders["month"]= orders["date_datetime"].dt.month
orders["month_name"]= orders["date_datetime"].dt.month_name()

orders["day"]= orders["date_datetime"].dt.day
orders["day_date_day"]=  orders["date_datetime"].dt.day_name()
orders["hour"]= orders["date_datetime"].dt.hour


orders["Dateof_week"]= orders["date_datetime"].dt.dayofweek
orders["Date_is_wknd"]= np.where(orders["Dateof_week"].isin(["Sunday", "Saturday"]),1,0)

orders["quarters"]= orders["date_datetime"].dt.quarter
```

```python
orders.drop(columns=["product_id", "city_id", "orders"], inplace=True)
orders.head(10)
```
